```python
a = open("trueEventTypePopln.txt")
import json
```

```python
etdict = {"NER": dict(), "nouns": dict()}
popdict = {"NER": dict(), "nouns": dict()}
i = 0
for l in a:
    j = json.loads(l.decode("utf-8"))
    #print j.keys()
    if "trueEventType" not in j:
        i += 1
        continue
        
    for feed in j["derivedFrom"]["derivedTws"]:
        for t in feed["BasisEnrichment"]["entities"]:
            if t["neType"].lower() != "location":
                etdict["NER"].setdefault(j["trueEventType"], [])
                etdict["NER"][j["trueEventType"]].append(t["expr"])
                popdict["NER"].setdefault(j["truePopulation"], [])
                popdict["NER"][j["truePopulation"]].append(t["expr"])
                
        for t in feed["BasisEnrichment"]["tokens"]:
            if "NOUN" in t["POS"]:
                etdict["nouns"].setdefault(j["trueEventType"], [])
                etdict["nouns"][j["trueEventType"]].append(t["lemma"])
                popdict["nouns"].setdefault(j["truePopulation"], [])
                popdict["nouns"][j["truePopulation"]].append(t["lemma"])
```

```python
with open("./eventTypeDist.txt", "w") as out:
    out.write(json.dumps(etdict, ensure_ascii=False).encode("utf-8"))
```

```python
with open("./popDist.txt", "w") as out:
    out.write(json.dumps(popdict, ensure_ascii=False).encode("utf-8"))
```
